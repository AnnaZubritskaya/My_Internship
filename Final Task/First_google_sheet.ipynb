```python
data=pd.read_csv('D:\df_task.csv')

grouped_data = data.groupby('Country')['Name of University'].apply(list).reset_index()

new_df = pd.DataFrame({'Country': grouped_data['Country'], 'Universities': grouped_data['Name of University']})

new_df
```

```python
list_count=[]

for i in new_df.index:
    list_univer=list(new_df.loc[i,'Universities'])
    list_count.append(len(list_univer))

new_df['Number of universities']=list_count
new_df
```

```python
x=new_df['Country']
y=new_df['Number of universities']

plt.barh(x,y)

plt.show()
```

```python
x=new_df['Country']
y=new_df['Number of universities']

plt.bar(x,y)
plt.xticks(rotation=45)
plt.show()
```

```python
x=new_df['Country']
y=new_df['Number of universities']

plt.pie(y, labels=x)

plt.show()
```

```python
x=new_df['Country']
y=new_df['Number of universities']

sorted_data = sorted(zip(y, x))

sorted_y_values, sorted_countries = zip(*sorted_data)

plt.barh(sorted_countries, sorted_y_values)

plt.show()
```
