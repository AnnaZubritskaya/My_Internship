```python
data=pd.read_excel('D:\proper_view.xlsx')

new_df=pd.DataFrame({'University': data['Name of University'], 'Website': data['Website Uni'], 'Website TTO': data['Website TTO']})
new_df
```

```python
contact_dict = {
    'Name': [],
    'Position': [],
    'Address': [],
    'Email': [],
    'Phone': [],
    'Link': []
}

for index, row in data.iterrows():
    contact_info = row['Name of contact point']
    
    info_dict = {
        'Name': np.nan,
        'Position': np.nan,
        'Address': np.nan,
        'Email': np.nan,
        'Phone': np.nan,
        'Link': np.nan
    }
    
    if not pd.isnull(contact_info):
        lines = contact_info.split('\n')        
        for line in lines:
            if line.startswith('Name:'):
                info_dict['Name'] = line.replace('Name: ', '')
            elif line.startswith('Position:'):
                info_dict['Position'] = line.replace('Position: ', '')
            elif line.startswith('Email:') or line.startswith('E-mail:'):
                info_dict['Email'] = line.replace('Email: ', '')
            elif line.startswith('Phone:'):
                info_dict['Phone'] = line.replace('Phone: ', '')
            elif line.startswith('https://www.linkedin.com'):
                info_dict['Link'] = line
            elif line.startswith('Address:'):
                info_dict['Address'] = line.replace('Address: ', '')
    
    contact_dict['Name'].append(info_dict['Name'])
    contact_dict['Position'].append(info_dict['Position'])
    contact_dict['Address'].append(info_dict['Address'])
    contact_dict['Email'].append(info_dict['Email'])
    contact_dict['Phone'].append(info_dict['Phone'])
    contact_dict['Link'].append(info_dict['Link'])

contact_dict
```

```python
new_df['Name of contact point']=contact_dict['Name']
new_df['Position of contact point']=contact_dict['Position']
new_df['Adress of contact point']=contact_dict['Address']
new_df['Email of contact point']=contact_dict['Email']
new_df['Phone of contact point']=contact_dict['Phone']
new_df['Link of contact point']=contact_dict['Link']
new_df
```
